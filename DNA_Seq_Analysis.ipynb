```python
from google.colab import files
from Bio.SeqUtils import gc_fraction




# Upload your FASTA file
uploaded = files.upload()
fasta_file = list(uploaded.keys())[0]

# Read all sequences
records = list(SeqIO.parse(fasta_file, "fasta"))
print(f"Number of sequences in file: {len(records)}")

# Example: first sequence
record = records[0]
dna_seq = record.seq
print("\nSequence ID:", record.id)
print("Sequence length:", len(dna_seq))

# Nucleotide frequency
freq = {base: dna_seq.count(base) for base in "ATGC"}
print("\nNucleotide frequency:", freq)


# GC / AT content
gc_content = gc_fraction(dna_seq) * 100
at_content = 100 - gc_content
print(f"GC content: {gc_content:.2f}%")
print(f"AT content: {at_content:.2f}%")




# Motif search (ATG)
motif = "ATG"
positions = [i+1 for i in range(len(dna_seq)) if dna_seq.startswith(motif, i)]
print(f"\nMotif '{motif}' found at positions:", positions)

# ORF detection
stop_codons = ["TAA", "TAG", "TGA"]
orfs = []
for i in range(len(dna_seq)-2):
    if dna_seq[i:i+3] == "ATG":
        for j in range(i, len(dna_seq)-2, 3):
            codon = dna_seq[j:j+3]
            if codon in stop_codons:
                orfs.append((i+1, j+3))
                break
print("\nNumber of ORFs found:", len(orfs))
if orfs:
    print("First 5 ORFs positions (start-end):", orfs[:5])

# Translation
protein = dna_seq.translate(to_stop=True)
print("\nProtein sequence (first 50 aa):", protein[:50])

# Amino acid composition
aa_counts = Counter(str(protein))
print("\nAmino acid composition (top 10):", aa_counts.most_common(10))

```
